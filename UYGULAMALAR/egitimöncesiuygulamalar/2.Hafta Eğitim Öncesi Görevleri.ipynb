YetGen liderlerinden Berkcan, Emirhan ve Eslem, Ramazan Bayramı'nda topladıkları şekerlerini bir araya getirmeyi ve kendi aralarında eşit olarak paylaşmayı kabul ettiler.
Eşit paylaşım yapıldıktan sonra kalan şekerler parçalanacak. Örneğin, eğer toplamda 91 şekerleri varsa , her biri 30 tane alıp kalan 1 tane şekeri parçalayacaklar.
Kaç tane şekeri parçalayacaklarını hesaplayan bir Python kodu yazınız.

````python
# YetGen liderlerinin topladığı şekerlerin sayılarını tanımlıyoruz.
berkcan_sayisi = 40
emirhan_sayisi = 30
eslem_sayisi = 21

# Toplam şeker sayısını hesaplıyoruz.
toplam_sayi = berkcan_sayisi + emirhan_sayisi + eslem_sayisi

# paylştıktan sonra kalan şekeri hesaplıyoruz
kalan = toplam_sayi % 3

# Eğer kalan şeker varsaeşit olarak paylaşıyoruz
if kalan != 0:
    parcalanacak_sayi = 3 - kalan
else:
    parcalanacak_sayi = 0

# Syazdırıyoruz.
print("Parçalanacak şeker sayısı:", parcalanacak_sayi)
````
